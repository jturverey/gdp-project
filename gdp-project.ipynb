It also explores the potential correlation between GDP and life expectancy for these six countries.  
The dataset spans the timeframe of 2000-2015.

## Project Imports

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Import dataset to be used in project.
gdp = pd.read_csv('gdp_life_expectancy.csv')
```

```python
# First 10 records of dataset
gdp.head(10)
```

## Statistics

```python
# Summary statistics for dataset
gdp.describe()
```

```python
# Distribution of life expectancy
sns.displot(gdp['Life Expectancy (years)'], bins=range(44, 83, 2))
plt.xticks(range(44, 82, 4))
plt.show()
```

We begin by looking at a distribution of life expectancy values for the entire dataset. The distribution is significantly left-skewed, with most values falling into the range of 74-80 years.

## Visualizations

```python
# Barchart showing average life expectancy in six countries
sns.barplot(data=gdp, x='Country', y='Life Expectancy (years)', palette='muted', ci=None)
plt.xlabel('Country', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)
plt.show()
```

The average life expectancy for most countries is very similar, generally falling into the range of 74-80 years. Zimbabwe has a significantly lower average life expectancy&mdash;approximately 50 years. 

```python
# Linechart showing how life expectancy has changed over time.
sns.lineplot(data=gdp, x='Year', y='Life Expectancy (years)', hue='Country')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)
plt.legend(loc=4, fontsize=9)
plt.show()
```

The life expectancy for most countries has increased at a similar, steady rate from 2000-2015. Zimbabwe however, began with a life expectancy which was decreasing, but then began to increase in 2004. By 2006, the life expectancy in Zimbabwe began to increase sharply. 

```python
# Linechart showing how GDP has changed over time.
sns.lineplot(data=gdp, x='Year', y='GDP (billions)', hue='Country')
plt.ticklabel_format(axis='y', style='plain')
plt.xlabel('Year', fontsize=12)
plt.ylabel('GDP (billions)', fontsize=12)
plt.legend(loc=2, fontsize=9)
plt.show()
```

In this chart, we see the USA and China stand out the most, with GDP generally increasing at significant rates throughout 2000-2015. The rate at which GDP increases in China becomes noticeably sharp beginning in 2006.  

```python
# Scatterplot showing a possible correlation between GDP and life expectancy.
sns.scatterplot(data=gdp, x='GDP (billions)', y='Life Expectancy (years)', hue='Country')
plt.xlabel('GDP (billions)', fontsize=12)
plt.ylabel('Life Expectancy (years)', fontsize=12)
plt.show()
```

In the USA, China, and Germany, we can see that GDP and life expectancy have a positive correlation. However, the other countries (Zimbabwe, Mexico, and Chile) also show an increasing life expectancy, while showing little or no increase in GDP.

## Conclusions

This dataset shows life expectancy to increase over time for all six countries. It increases at a particualrly high rate for Zimbabwe, going from 44 years in 2004, to over 60 years by 2015.  
GDP and life expectancy show a correlation for many of the countries, but it is apparent that life expectancy also increases in countries that do not have an increasing rate of GDP. As a result, we can conclude that even though GDP and life expectancy are positively correlated in many instances, the life expectancy of a country is not dependent on its GDP growth rate.
